```python
from os.path import splitext
from os import system
```

```python
def modify_filename(filename):
    import string
    valid_chars = "-_.() %s%s" % (string.ascii_letters, string.digits)
    return ''.join(c for c in filename if c in valid_chars)
```

```python
n = 241
for item in videolist[n:]:
    n += 1
    try:
        print (item)
        yt = YouTube(item)
        print (n, yt.title)
        if int(yt.length) > 300:
            continue
        audio = yt.streams.filter(only_audio=True).all()[0]
        print ('Downloading ' + str(n) + '...')
        filename = audio.default_filename
        new_filename = "%d_%s%s" % (n, modify_filename(splitext(filename)[0]), ".wav")
        path = './audios/alphago_news_us/'
        audio.download('./')
        system('ffmpeg -i "%s" "%s"' % (filename, path + new_filename))
        system('rm "%s"' % filename)
    except (VideoUnavailable, RegexMatchError, IndexError, UnboundLocalError):
        continue
```
