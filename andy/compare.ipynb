```python
def gen_params(named_entity=None, channel_id=None):
    return {
        'q': 'alphago %s' % named_entity,
        'part': 'snippet',
        'key': youtube_api_key,
        'type': 'video',
        'maxResults': 50,
        'channelId': channel_id
    }

def get_num_results(named_entity, channel_id):
    r = requests.get(url, params=gen_params(named_entity=named_entity, channel_id=channel_id))
    results_json = r.json()
    return results_json['pageInfo']['totalResults']
```

```python
entities = []
plain_counts = []
cgtn_counts = []
with open('./wordcounts/named_entity.count') as f:
    lines = f.read().splitlines()
with open('./wordcounts/named_entity_search_results.count', 'w+') as f:
    for line in lines:
        try:
            entity = line.split()[1]
            plain_count = get_num_results(entity, None)
            cgtn_count = get_num_results(entity, cgtn_channel_id)
            f.write('\t'.join([entity, str(plain_count), str(cgtn_count)]) + '\n')
        except IndexError:
            continue
```

```python
with open('./wordcounts/named_entity_search_results.count') as f:
    lines = f.read().splitlines()
    for line in lines:
        content = line.split('\t')
        entities.append(content[0])
        plain_counts.append(int(content[1]))
        cgtn_counts.append(int(content[2]))
```

```python
plain_sum = sum(plain_counts)
plain_counts = [round(p / plain_sum * 100, 2 )for p in plain_counts]
```
